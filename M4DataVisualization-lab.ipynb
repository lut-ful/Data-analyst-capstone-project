## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
qry="select convertedComp from master"
df=pd.read_sql_query(qry,conn)
df.head(5)
```

```python
import matplotlib.pyplot as plt
df.plot(kind='hist',figsize=(15,10))
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
qry="select age from master"
df=pd.read_sql_query(qry,conn)
df.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
qry="select age,workweekhrs from master"
df=pd.read_sql_query(qry,conn)
df.columns
```

```python
df.plot(kind='scatter',x='Age',y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
df=pd.read_sql_query("select workweekhrs,coderevhrs,age from master",conn)
df.head(5)
```

```python
# your code goes here
plt.scatter(x=df['WorkWeekHrs'],y=df['CodeRevHrs'],s=df['Age'],alpha=.5)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
qry = """SELECT 
DatabaseDesireNextYear as Database, 
count(*) as No_of_respondents 
FROM DatabaseDesireNextYear 
GROUP BY DatabaseDesireNextYear 
ORDER BY No_of_respondents DESC LIMIT 5"""
df=pd.read_sql_query(qry,conn)
df
```

```python
plt.pie(x=df['No_of_respondents'],labels=df['Database'],autopct="%11.1f%%")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
qry=""" select workweekhrs,coderevhrs, age from master where age between 30 and 35"""
df=pd.read_sql_query(qry,conn)
df.head()
```

```python
for_stack_chart=df.groupby('Age')[['WorkWeekHrs','CodeRevHrs']].median()
```

```python
for_stack_chart
```

```python
for_stack_chart.plot(kind='bar',stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
qry="""select age,convertedcomp from master where age between 45 and 60"""
df=pd.read_sql_query(qry,conn)
df
```

```python
df=df.groupby('Age')['ConvertedComp'].median()
```

```python
df.plot()
```
